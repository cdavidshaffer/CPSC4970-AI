```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_breast_cancer
from IPython.core.display import display
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
df = load_breast_cancer(as_frame=True)['frame']
display(df.info())
display(df.head())
display(df.describe())
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df['target'], test_size=0.33)
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(X_train.corr(), annot=True, fmt='.2f')
```

```python
cols_to_remove = ['mean perimeter', 'mean area', 'worst radius', 'worst perimeter', 'worst area',
                  'worst texture', 'worst compactness', 'mean concavity', 'worst concave points',
                  'perimeter error', 'area error']
X_train = X_train.drop(columns=cols_to_remove)
X_test = X_test.drop(columns=cols_to_remove)
plt.figure(figsize=(16,16))
sns.heatmap(X_train.corr(), annot=True, fmt='.2f')
```

```python
classifier = GaussianNB().fit(X_train, y_train)
pred_train = classifier.predict(X_train)
pred_test = classifier.predict(X_test)
plot_confusion_matrix(classifier, X_train, y_train)
print("Training accuracy: ", accuracy_score(y_train, pred_train))
print("Testing accuracy: ", accuracy_score(y_test, pred_test))
```
