```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from IPython.core.display import display
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelBinarizer

d = load_iris()
X_train, X_test, y_train, y_test = train_test_split(d.data, d.target, test_size=0.33)
print(X_train[:10,:])
print(X_train.shape)
print(y_train[:10])
```

```python
input_encoder = StandardScaler()
X_train_trans = input_encoder.fit_transform(X_train)
X_test_trans = input_encoder.transform(X_test)
print(X_train_trans[:10,:])
```

```python
output_encoder = LabelBinarizer()
y_train_trans = output_encoder.fit_transform(y_train)
y_test_trans = output_encoder.transform(y_test)
print(y_train_trans[:10])
```

```python
from keras.callbacks import EarlyStopping
from keras.losses import CategoricalCrossentropy
from keras.layers import Dense
from keras import Input, activations
from keras.models import Sequential

model = Sequential([
    Input(shape=X_train_trans.shape[1]),
    Dense(4, activation=activations.sigmoid),
    Dense(y_train_trans[0].shape[0], activation=activations.softmax)
])
model.summary()
callback = EarlyStopping(monitor='loss', patience=20)
model.compile(optimizer='adam', loss=CategoricalCrossentropy(), metrics=['accuracy'])
model.fit(X_train_trans, y_train_trans, epochs=1000, callbacks=[callback])
```

```python
from sklearn.metrics import accuracy_score

pred_train_cat = output_encoder.inverse_transform(model.predict(X_train_trans))
pred_test_cat = output_encoder.inverse_transform(model.predict(X_test_trans))
print("Training accuracy:", accuracy_score(y_train, pred_train_cat))
print("Testing accuracy:", accuracy_score(y_test, pred_test_cat))
```
